# Introduction

Pairs trading is a market-neutral strategy that seeks to exploit relative mispricings between two co-integrated assets. The main idea is that the two assets usually move together in a predictable way. If their prices temporarily move too far apart, we expect them to return to their normal relationship. This creates a chance to buy the cheaper one and sell the expensive one, making a profit when they realign.

# Step 1: Data Collection and Cointegration Analysis

This step identifies asset pairs suitable for pairs trading by following these steps:

### **1. Download Historical Data**
- Retrieves **10 years** of price data from Yahoo Finance.
- Uses **adjusted close prices** for accuracy.

### **2. Apply Log Transformation**
- Converts prices to **log scale** to stabilize variance.

### **3. Test for Stationarity (ADF Test)**
- Applies the **Augmented Dickey-Fuller test (ADF)** to check if each assetâ€™s price series is stationary.

### **4. Test for Cointegration (Engle-Granger)**
- Runs a **linear regression** between the two assets.
- Extracts the **residuals** and applies the **ADF test**.
- If residuals are **stationary**, the pair is **cointegrated** and suitable for trading.

### **5. Analyze Multiple Asset Pairs**
- Evaluates pairs like **gold & silver, Visa & Mastercard, Meta & Microsoft**.
- Only **cointegrated pairs** proceed to the trading strategy.

**Outcome:**  
Identifies pairs with a **statistically validated relationship**, ensuring a solid foundation for the trading model.

### **Engle-Granger Cointegration Test: Interpretation**  

- **p-value (0.0500):** Right at the **5% threshold**, indicating weak evidence of cointegration.  

**Conclusion:** The pair shows marginal cointegration.

# **Conclusion**  

This project implemented a **pairs trading strategy** using the **co-integration approach** with a **Kalman Filter** to dynamically estimate the hedge ratio. The strategy was backtested on **10 years of historical data** using **Shell (SHEL) and Valero (VLO)** as the selected pair.  

- **Cointegration Confirmed**: The Engle-Granger and Johansen tests provided evidence of a **long-term relationship** between SHEL and VLO.
